```python
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_fence = Q1 - 1.5 * IQR
        upper_fence = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_fence) & (df[col] <= upper_fence)]
    return df

# ✅ Only use numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
df = remove_outliers_iqr(df, numeric_cols)
```

```python
# Vif library will give me a VIF value for every column
# If any column is having VIF value more than 6 please ignore that column

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df.info()
```

```python
col_list=[]

for i in df.columns:
  if ((df[i].dtype !="object") & (i!="price")):  # Apply VIf only on Numerical columns
    col_list.append(i)
```

```python
col_list
```

```python
X=df[col_list]   # Store all column in a variable X
vif_data = pd.DataFrame()  # Create a datafarem to store column name and its VIF value
vif_data["Column Name"] =X.columns
vif_data["VIF value"]=[variance_inflation_factor(X.values,i)for i in range(len(X.columns))]  # Finding the VIF value for columns
print(vif_data)
```
